```python
# 0.02 in latitude is equivalent to 2km
def is_within_proximity(row1, row2, tolerance=0.02):
    """Check if two locations are within the specified tolerance in lat/lon."""
    lat_diff = abs(row1['latitude'] - row2['latitude'])
    lon_diff = abs(row1['longitude'] - row2['longitude'])
    
    return lat_diff <= tolerance and lon_diff <= tolerance
```

lag_1: 
- Count of outflows from the previous hour.
- Purpose: Value of the Outflow_Count at the previous time point might have an influence on the current value. This is useful when there's autocorrelation in the data, meaning past values are predictive of future values.

rolling_avg_3
- Averaging the values of the target variable over the last 3 hours.
- Purpose: smooth out short-term fluctuations and capture longer-term trends in the data, helps to reduce the noise and makes the model focus on the underlying trend.
