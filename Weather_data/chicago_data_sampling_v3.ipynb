```python
def reduce_size(df):
    '''
    Reduce pandas dataframe memory usage by optimising datatypes
    '''
    mem_before_reduction = df.memory_usage().sum()/1024**2
    print(f'Memory usage before optimising datatypes: {mem_before_reduction:.0f} MB')

    for col in df.columns:

        if str(df[col].dtype)[:3] == 'int':
            min_val = df[col].min()
            max_val = df[col].max()
            
            if min_val > np.iinfo(np.int8).min and max_val < np.iinfo(np.int8).max:
                df[col] = df[col].astype(np.int8)
            elif min_val > np.iinfo(np.int16).min and max_val < np.iinfo(np.int16).max:
                df[col] = df[col].astype(np.int16)
            elif min_val > np.iinfo(np.int32).min and max_val < np.iinfo(np.int32).max:
                df[col] = df[col].astype(np.int32)
            elif min_val > np.iinfo(np.int64).min and max_val < np.iinfo(np.int64).max:
                df[col] = df[col].astype(np.int64)
        elif str(df[col].dtype)[:5] == 'float':
            min_val = df[col].min()
            max_val = df[col].max()
            
            if min_val > np.finfo(np.float16).min and max_val < np.finfo(np.float16).max:
                df[col] = df[col].astype(np.float16)
            elif min_val > np.finfo(np.float32).min and max_val < np.finfo(np.float32).max:
                df[col] = df[col].astype(np.float32)
            else:
                df[col] = df[col].astype(np.float64)
    
    
    mem_after_reduction = df.memory_usage().sum()/1024**2
    print(f'Memory usage after optimising datatyes: {mem_after_reduction:.0f} MB')
    return df
    
```

```python
# Strategies

## 1. Drop latitudes, longtidues and retain centroids
## 2. Keep timestamps as objects
## 3. Convert id from object to serial numbers
```

```python
station_list = df[['pickup_centroid_location']].drop_duplicates()

station_list['longitude'] = station_list['pickup_centroid_location'].str.extract(r'POINT \((-?\d+\.\d+)')[0].astype(float)
station_list['latitude'] = station_list['pickup_centroid_location'].str.extract(r'(-?\d+\.\d+)\)$')[0].astype(float)
```

```python
def get_nearest_station(lat, lon):
    location = Point(lat, lon)
    stations = Stations().nearby(lat, lon)
    station = stations.fetch(1)  # Fetch the nearest station
    return station.index[0] if not station.empty else '72530'

station_list['station_id'] = station_list.apply(lambda row: get_nearest_station(row['latitude'], row['longitude']), axis=1)

# Display the DataFrame with station IDs
station_list.head()
```

```python
list_of_stations = pd.DataFrame(list(set(station_list['station_id'])),columns=['station_id'])
station_list = station_list.drop(['longitude','latitude'], axis=1)
gc.collect()
```

```python
df = df.merge(station_list,on='pickup_centroid_location', how='left')
df['station_id'] = df['station_id'].replace('KCGX0', '72530')
```
