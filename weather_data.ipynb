## Hourly Data

```python
from datetime import datetime
import pandas as pd
from meteostat import Hourly

# Set time period
start = datetime(2014, 1, 1)
end = datetime(2024, 5, 31, 23, 59)

# List to store hourly weather data DataFrames
hourly_data_list = []

# Function to fetch hourly weather data
def fetch_hourly_data(station_id, city, state):
    try:
        data = Hourly(station_id, start, end)
        data = data.fetch()
        data.index = pd.to_datetime(data.index)
        data['City'] = city
        data['State'] = state
        data['Full Date'] = data.index
        data['Hour'] = data.index.hour
        data['Day'] = data.index.day
        data['Month'] = data.index.month
        data['Year'] = data.index.year
        return data
    except Exception as e:
        print(f"Failed to fetch data for station {station_id} in {city}, {state}: {e}")
        return None

# Assuming geocoding DataFrame has 'Station ID', 'City', and 'State' columns
# Replace this with your actual DataFrame
geocoding = pd.read_csv('cities_with_station_ids.csv')

# Loop through each row in the geocoding DataFrame and fetch hourly weather data
for index, row in geocoding.iterrows():
    station_id = row['Station ID']
    city = row['City']
    state = row['State']
    hourly_data = fetch_hourly_data(station_id, city, state)
    if hourly_data is not None:
        hourly_data_list.append(hourly_data)

# Concatenate all the hourly DataFrames
if hourly_data_list:
    hourly = pd.concat(hourly_data_list, ignore_index=True)
    # Save the hourly DataFrame to a CSV file
    hourly.to_csv('hourly_weather_data.csv', index=False)
    print("Hourly weather data has been saved to hourly_weather_data.csv")
else:
    print("No data fetched.")
```

```python
import pandas as pd
from astral.sun import sun
from astral import LocationInfo
from datetime import datetime, timedelta

def calculate_solar_events(lat, lon, date, city, state):
    location = LocationInfo(name=city, region=state, timezone="UTC", latitude=lat, longitude=lon)
    solar_events_list = []
    
    try:
        s = sun(location.observer, date=date, tzinfo='UTC')
        daylight_duration = (s['sunset'] - s['sunrise']).seconds / 3600.0
        for hour in range(24):
            solar_events_list.append({
                'City': city,
                'State': state,
                'Latitude': lat,
                'Longitude': lon,
                'Full Date': datetime.combine(date, datetime.min.time()) + timedelta(hours=hour),
                'Dawn': s['dawn'].time(),
                'Sunrise': s['sunrise'].time(),
                'Noon': s['noon'].time(),
                'Sunset': s['sunset'].time(),
                'Dusk': s['dusk'].time(),
                'Daylight Hours': daylight_duration
            })
    except ValueError as e:
        print(f"Skipping date {date} for {city}, {state}: {e}")
    
    return solar_events_list

# Ensure 'Full Date' is in datetime format
df['Full Date'] = pd.to_datetime(df['Full Date'])

# Get unique dates and locations from the weather hourly DataFrame
unique_dates = df['Full Date'].dt.date.unique()
unique_locations = df[['City', 'State', 'Latitude', 'Longitude']].drop_duplicates()

# List to store solar events data
solar_events_list = []

# Loop through each unique location and calculate solar events data
for index, row in unique_locations.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    city = row['City']
    state = row['State']
    for date in unique_dates:
        solar_events = calculate_solar_events(lat, lon, date, city, state)
        solar_events_list.extend(solar_events)

# Convert the list of results to a DataFrame
solar_events_df = pd.DataFrame(solar_events_list)

# Merge the solar events data with the weather hourly data
merged_df = df.merge(solar_events_df, on=['City', 'State', 'Full Date'], how='left')
```
